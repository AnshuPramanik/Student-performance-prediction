# Data Cleaning
This notebook performs data cleaning operations on the student performance dataset.

## Import Libraries and Load Dataset

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Student_Performance.csv')

# Display basic info
print("Dataset shape:", df.shape)
print("Columns:", df.columns.tolist())
print("Missing values:\n", df.isnull().sum())
print("Duplicates:", df.duplicated().sum())
```

## Remove Duplicates

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
print("After removing duplicates:", df.shape)
```

## Rename Columns and Drop Unnecessary Columns

```python
# Rename subject columns
df.rename(columns={
    'math_score': 'computer_network_score',
    'science_score': 'operating_system_score',
    'english_score': 'daa_score'
}, inplace=True)

# Drop student_id as it's not useful for prediction
df.drop('student_id', axis=1, inplace=True)

print("Columns renamed and student_id dropped")
```

## Check Data Types

```python
# Check data types
print("Data types:\n", df.dtypes)
```

## Save Cleaned Dataset

```python
# Save cleaned dataset
df.to_csv('cleaned_dataset.csv', index=False)

print("Cleaned dataset saved as 'cleaned_dataset.csv'")
```
